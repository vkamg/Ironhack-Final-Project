```python
import pandas as pd
import re
import ast
```

```python
recipe_df = pd.read_json('../data/raw/web_scraping_recipes_recetasgratis.json')

recipe_df.head()
```

```python
def clean_receta(column):
    clean_string = re.sub("Recetas? de ", "", column)
    return (clean_string)

recipe_df["recipe_name"] = recipe_df.apply(lambda x: clean_receta(x['recipe_name']), axis=1)

recipe_df["category"] = recipe_df.apply(lambda x: clean_receta(x['category']), axis=1)
```

```python
recipe_df["category"] = recipe_df["category"].str.lower()

category_list = recipe_df["category"].unique()

category_list
```

```python
def main_category_clasification(column):
    if column == "postres":
        return "postre"
    elif column == "pan y bollería":
        return "postre"
    elif column == "aperitivos y tapas":
        return "tapas"
    else:
        return "principal"

recipe_df["main_category"] = recipe_df.apply(lambda x: main_category_clasification(x["category"]), axis=1)
```

```python
recipe_df['time_preparation'].fillna('missing', inplace=True)
```

```python
def object_tonumber(column):
    if column == "missing":
        return None
    elif re.search("h", column) is not None:
        num_h = re.findall("(\d+)h", column)
        if re.search("m", column) is not None:
            num_min = re.findall("(\d+)m", column)
            total = int(num_h[0]) * 60 + int(num_min[0])
            return total
        else:
            total = int(num_h[0]) * 60
            return total
    else:
        object_split = column.split("m")
        result = int(object_split[0])
        return result
```

```python
recipe_df["time_preparation(min)"] = recipe_df.apply(lambda x: object_tonumber(x["time_preparation"]), axis=1)
```

```python
recipe_df["time_preparation(min)"].value_counts()
```

```python
def empty_lists(column):
    if len(column) == 0:
        column.append("no notes")
        return column
    else:
        return column

recipe_df["more_info"] = recipe_df.apply(lambda x: empty_lists(x["more_info"]), axis=1)
```

```python
def check_notes_values(column, df):
    notes_list = []
    for x in df[column]:
        notes_list.append(x)
    all_notes = [val for sublist in notes_list for val in sublist]
    return list(set(all_notes))

unique_notes = check_notes_values("more_info", recipe_df)

unique_notes
```

```python
intolerances_cols = ["gluten_free", "egg_free", "sucrose_fructose_free", "low_sodium", "lactose_free"]
```

```python
def gluten_free(intolerances):
    if 'Receta sin gluten' in intolerances:
        return 1
    else:
        return 0

recipe_df["gluten_free"] = recipe_df.apply(lambda x: gluten_free(x["more_info"]), axis=1)
```

```python
def egg_free(intolerances):
    if 'Receta sin huevo' in intolerances:
        return 1
    else:
        return 0

recipe_df["egg_free"] = recipe_df.apply(lambda x: egg_free(x["more_info"]), axis=1)
```

```python
def sucrose_fructose_free(intolerances):
    if 'Receta sin azúcar' in intolerances:
        return 1
    else:
        return 0
    
recipe_df["sucrose_fructose_free"] = recipe_df.apply(lambda x: sucrose_fructose_free(x["more_info"]), axis=1)
```

```python
def low_sodium(intolerances):
    if 'Receta sin sal' in intolerances:
        return 1
    else:
        return 0

recipe_df["low_sodium"] = recipe_df.apply(lambda x: low_sodium(x["more_info"]), axis=1)
```

```python
def lactose_free(intolerances):
    if 'Receta sin lactosa' in intolerances:
        return 1
    else:
        return 0

recipe_df["lactose_free"] = recipe_df.apply(lambda x: lactose_free(x["more_info"]), axis=1)
```

```python
def ingredients_tolist(ingred_column):
    ingredients_list = ingred_column.split("\n")
    return ingredients_list

recipe_df["ingredients_list"] = recipe_df.apply(lambda x: ingredients_tolist(x['ingredients']), axis=1)
```

```python
recipe_df
```

```python
def clean_ingredients(ingred_column):
    #ingred = ast.literal_eval(ingred_column)
    ingredients_list = []
    for ing in ingred_column:
        clean_ingred = re.sub("para.*$|\d(.*) de |\d |\([^)]*\)| el peso de.*$| medi[a-z][a-z]?[a-z]? con.*$|yemas? de |claras? de |½ (.*) de |½ |¼ (.*) de |¼ ", "", ing)
        ingredients_list.append(clean_ingred)
        if clean_ingred == "sal y pimienta":
            x = clean_ingred.split(" y ")
            ingredients_list.append(x[0])
            ingredients_list.append(x[1])
            ingredients_list.remove(clean_ingred)
    ingredients_list = [i for i in ingredients_list if i]
    return ingredients_list
```

```python
recipe_df["ingredients_list"] = recipe_df.apply(lambda x: clean_ingredients(x['ingredients_list']), axis=1)
```

```python
recipe_df
```

```python
def num_ingred(column):
    return len(column)

recipe_df["num_ingredients"] = recipe_df.apply(lambda x: num_ingred(x["ingredients_list"]), 
                                                         axis=1)
```
