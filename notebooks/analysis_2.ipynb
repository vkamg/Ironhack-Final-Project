```python
def input_list(string):
    input_list = []
    str_split = string.split(", ")
    for x in str_split:
        input_list.append(x)
    return input_list

input_ingred = input_list(input_ing)
```

```python
def fuzzy_columns(column, input_string):
    return fuzz.token_set_ratio(column, input_string)
```

```python
for ingred in input_ingred:
    recipe_df[ingred] = recipe_df.apply(lambda x: fuzzy_columns(x["ingredient"], ingred), axis=1)
```
